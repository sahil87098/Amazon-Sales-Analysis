# Exploratory Data Analysis

```python
data.columns
```

### Size

```python
ax = sns.countplot(x='Size', data=data)
# Using seaborn to check size
```

```python
ax=sns.countplot(x='Size',data=data)

for bars in ax.containers:
    ax.bar_label(bars)
```

The graph shows that most people buy "M-size".

Now using groupby() function to group data based on one or more columns in a DataFrame.

```python
data.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
quant = data.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
sns.barplot(x='Size',y='Qty',data=quant)
```

This graph shows that the most of the Qty buys "M-size" in sales.

# Histogram on Size

```python
data['Size'].hist()
```

```python
data['Category'] = data['Category'].astype(str)
column_data = data['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data,bins=30,edgecolor='Black')
plt.xticks(rotation=90)
plt.show()
# Converted Category to string
# Made a variable called column_data
# "bins" is used for distance between bars
```

This graph shows that the most are buying "T-shirts".

# Pie Chart for checking B2B Data

```python
B2B_Check = data['B2B'].value_counts()

plt.pie(B2B_Check, labels=B2B_Check.index, autopct='%1.1f%%')
# plt.axis('equal')
plt.show()
```

This chart shows that, 99.2% buyers are retailers and 0.8% are B2B buyers.

# Scatter Plot

Using scatter plot to find how many data and how much there are different types of size.

```python
x_data = data['Category']
y_data = data['Size']

plt.scatter(x_data, y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=data, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()

# Plot count of cities by state
```

### Top 10 States

```python
top_10_states = data['ship-state'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.countplot(data=data[data['ship-state'].isin(top_10_states.index)], x= 'ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=45)
plt.show()
```

This above graph shows that most of the buyers are from "Maharashtra State".
